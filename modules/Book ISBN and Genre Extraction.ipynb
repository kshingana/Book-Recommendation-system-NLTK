```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests
import json
from IPython.core.display import Image
from isbntools.app import *
from time import perf_counter
#Books
#i_cols = ['isbn', 'book_title' ,'book_author','year_of_publication', 'publisher', 'img_s', 'img_m', 'img_l']
items = pd.read_csv('items_clean.csv')
items = items.drop("Unnamed: 0",axis=1)
df = items.copy()
#df["extracted_bookname"] = "NaN"
df["google_id"] = "NaN"
df["book_language"] = "NaN"
df["categories"] = ""
#df["previewLink"] = "NaN"
#df["description"] = "NaN"
df["extracted_isbn"] = "NaN"
#.......................................................................................................................
t1_start = perf_counter()
error_list = []
#.......................................................................................................................
start = int(input("Enter START number : "))
end = int(input("Enter END number : "))
t1_start = perf_counter()
for i in range(start,end):
    print(f"Counter = {i}")
    ISBN = df["ISBN"].iloc[i]
    try:
        try:
            get_isbn = df["ISBN"].iloc[i]
            url_1 = f"https://www.googleapis.com/books/v1/volumes?q=isbn:{get_isbn}"
            response1 = requests.get(url_1)
            data1 = response1.json()
            google_id = data1["items"][0]["id"]
            df["google_id"][i] = google_id
            book_laguage = data1["items"][0]["volumeInfo"]["language"]
            df["book_language"][i] = book_laguage
            categories = "".join(data1["items"][0]["volumeInfo"]["categories"])
            df["categories"][i] = categories
        except:
                book_name = df["Book-Title"].iloc[i]
                get_isbn = isbn_from_words(f"{book_name}")
                url_3 = f"https://www.googleapis.com/books/v1/volumes?q=isbn:{get_isbn}"
                response3 = requests.get(url_3)
                data3 = response3.json()
                google_id = df["items"][0]["id"]
                df["google_id"][i] = google_id
                categories = "".join(data3["items"][0]["volumeInfo"]["categories"])
                df["categories"][i] = categories
                book_laguage = data3["items"][0]["volumeInfo"]["language"]
                df["book_language"][i] = book_laguage
                df["extracted_isbn"][i] = get_isbn

        finally:
            print("........................................................................................................")
    except:
        print(f"inside outer except block......problem with {ISBN}")
        error_list.append(ISBN)
        print("............................................................................................................")

        
#................................................................................................        

df_extraction = df[start:end]
        
t1_stop = perf_counter()  
print("Elapsed time during the whole program in Min:",((t1_stop-t1_start)/60))
print("Errors : ",len(error_list))


df_extraction.to_csv(f"df_{start}to{end}.csv")
np.savetxt(f"error_log{start}to{end}.csv", 
           error_list,
           delimiter =", ", 
           fmt ='% s')
```

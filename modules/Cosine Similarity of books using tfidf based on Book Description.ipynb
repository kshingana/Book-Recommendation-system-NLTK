### Problem Statement_1 : Cosine Similarity of books using tfidf based on Book Description

Cosine similarity is a metric used to determine how similar the documents are irrespective of their size.


```python
from nltk.tokenize import word_tokenize
from gensim import models
import numpy as np
# Compute Cosine Similarity
from sklearn.metrics.pairwise import cosine_similarity
import os
```

```python
from gensim.matutils import softcossim
from gensim import corpora
import gensim.downloader as api
# Scikit Learn
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
```

```python
df = pd.read_csv("description_df_0to100.csv")
df.drop(columns = ["Unnamed: 0"],inplace=True)
df.head()
```

```python
# Define the documents

documents = []
for i in range(0,20):
    doc_i_ = df["description"].iloc[i]
    documents.append(doc_i_)
```

```python
documents
```

```python
count_vectorizer = CountVectorizer(stop_words='english',token_pattern=r'\w+')
count_vectorizer = CountVectorizer()
sparse_matrix = count_vectorizer.fit_transform(documents)
```

```python
doc_term_matrix = sparse_matrix.todense()
df = pd.DataFrame(doc_term_matrix, 
                  columns=count_vectorizer.get_feature_names())
df
```

```python
# Compute Cosine Similarity
from sklearn.metrics.pairwise import cosine_similarity
cosinesiml = (cosine_similarity(df, df))
```
